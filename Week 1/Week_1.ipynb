```python
# Histogram with Density Curve and Gap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
salary_cluster1 = np.random.normal(30000,3000,500)
salary_cluster2 = np.random.normal(80000,4000,500)
salary = np.concatenate([salary_cluster1, salary_cluster2])
plt.figure(figsize=(8,5))
sns.histplot(salary, bins = 30, kde = True, color = 'skyblue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
# Area chart
import matplotlib.pyplot as plt
months = ['Jan','Feb', 'Mar', 'Apr', 'May','Jun']
sales = [120, 130, 150, 140, 100, 80]
plt.figure(figsize = (8,5))
plt.fill_between(months, sales, alpha = 0.5)
plt.plot(months , sales, marker = 'o')
plt.title("Monthly Sales Area chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Normal Distribution Chart
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
x = np.linspace(-4,4,1000)
plt.figure(figsize=(8,5))
plt.plot(x, norm.pdf(x,0,1))
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```

# 5. Normal Distribution

Definition:
A normal distribution is a symmetric probability distribution centered around the mean.

Interpretation:
Most observations occur near the center, with fewer observations farther away.

Learning:
Many real-world datasets approximately follow a normal distribution.

```python
# Binomial Distribution
from scipy.stats import binom
import matplotlib.pyplot as plt
n = 20
p = 0.5
x = range(n+1)
y = binom.pmf(x, n, p)
plt.figure(figsize=(8,5))
plt.bar(x,y)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

# 6. Binomial Distribution

Definition:
The binomial distribution models the number of successes in a fixed number of independent trials.

Interpretation:
The highest probability occurs around 10 successes when n=20 and p=0.5.

Learning:
Binomial distributions are used in quality control, surveys, and experiments.

```python
# Poisson Distribution Chart
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
lam = 5
x = np.arange(0,20)
y = poisson.pmf(x, lam)
plt.figure(figsize=(8,5))
plt.bar(x,y)
plt.title("Poisson Distribution")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.show()
```

# 7. Poisson Distribution

Definition:
The Poisson distribution models the number of events occurring in a fixed interval.

Interpretation:
Events near λ=5 occur most frequently.

Learning:
Poisson distributions are commonly used for traffic, call arrivals, and accident modeling.

```python
# Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(42)
sample_means = []

for _ in range(1000):
    sample = np.random.uniform(0, 100, 30)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(8,5))
plt.hist(sample_means,bins=30,density=True,alpha=0.6)
mu = np.mean(sample_means)
sigma = np.std(sample_means)
x = np.linspace(min(sample_means),max(sample_means),200)
plt.plot(x, norm.pdf(x, mu, sigma),linewidth=2)
plt.title("Central Limit Theorem")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

# 8. Central Limit Theorem (Very Important)

Definition:
The Central Limit Theorem states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the original population distribution.

Why It Matters:
It allows statistical inference and hypothesis testing even when the original data is not normally distributed.

Interpretation:
Although samples are drawn from a uniform distribution, the sample means form a bell-shaped normal curve.

Learning:
The CLT is one of the most important concepts in statistics and machine learning because it justifies many statistical methods.
